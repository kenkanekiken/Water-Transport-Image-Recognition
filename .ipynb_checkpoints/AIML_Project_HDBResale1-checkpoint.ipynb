```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Data Loading

The HDB resale flat price dataset was loaded and inspected to understand its structure and attributes.


```python
df = pd.read_csv("./dataset/HDB_Resales_Flat_Price.csv")

print("Sample Data: ")
print(df.head())
print("\n==========================================================================================")

print("\nAnalytics: ")
print(df.describe())
print("\n==========================================================================================")

# Display rows with missing values on the DataFrame
print("\nNo of missing values: ")
print(df.isnull().sum())
print("\n==========================================================================================")

print("\nNo of duplicated values: ")
print(df.duplicated().sum())
print("\n==========================================================================================")
```

## 2. Data Cleaning

Duplicate records were removed to prevent biased learning and ensure data consistency.


```python
df = df.drop_duplicates()
print("No of duplicate values after removing duplicates:")
print(df.duplicated().sum())
```

## 3. Data Transformation

New features were created to better represent housing characteristics and improve predictive performance.


```python
df["month"] = pd.to_datetime(df["month"])
df["year"] = df["month"].dt.year

df["flat_age"] = df["year"] - df["lease_commence_date"]

def parse_remaining_lease(x):
    x = x.lower()
    years = int(x.split("year")[0].strip())
    months = 0
    if "month" in x:
        months = int(x.split("month")[0].split()[-1])
    return years + months / 12

df["remaining_lease_years"] = df["remaining_lease"].apply(parse_remaining_lease)
```

## 4. Exploratory Data Analysis (EDA)

EDA was performed to understand relationships between variables, price distribution, and potential outliers.


```python
fig, axes = plt.subplots(3, 3, figsize=(20, 14))
fig.suptitle("HDB Resale Flat Price â Comparative Analysis", fontsize=16)

# ===================== Row 1: Average Resale Price =====================
df.groupby("flat_type")["resale_price"].mean().sort_values() \
  .plot(kind="bar", ax=axes[0,0], color="steelblue", title="Avg Price by Flat Type")
axes[0,0].set_ylabel("Price ($)")
axes[0,0].tick_params(axis="x", rotation=45)

df.groupby("town")["resale_price"].mean().sort_values(ascending=False).head(10) \
  .plot(kind="bar", ax=axes[0,1], color="darkorange", title="Avg Price by Town (Top 10)")
axes[0,1].tick_params(axis="x", rotation=45)

df.groupby("storey_range")["resale_price"].mean().sort_index() \
  .plot(kind="bar", ax=axes[0,2], color="seagreen", title="Avg Price by Storey Range")
axes[0,2].tick_params(axis="x", rotation=45)

# ===================== Row 2: Transaction Count =====================
df["flat_type"].value_counts() \
  .plot(kind="bar", ax=axes[1,0], color="mediumpurple", title="Transaction Count by Flat Type")
axes[1,0].set_ylabel("Count")
axes[1,0].tick_params(axis="x", rotation=45)

df["town"].value_counts().head(10) \
  .plot(kind="bar", ax=axes[1,1], color="crimson", title="Transaction Count by Town (Top 10)")
axes[1,1].tick_params(axis="x", rotation=45)

df["storey_range"].value_counts().sort_index() \
  .plot(kind="bar", ax=axes[1,2], color="teal", title="Transaction Count by Storey Range")
axes[1,2].tick_params(axis="x", rotation=45)

# ===================== Row 3: Time / Model / Lease =====================
df.groupby("year")["resale_price"].mean() \
  .plot(kind="bar", ax=axes[2,0], color="royalblue", title="Avg Price by Year")
axes[2,0].set_ylabel("Price ($)")

df.groupby("flat_model")["resale_price"].mean().sort_values(ascending=False).head(10) \
  .plot(kind="bar", ax=axes[2,1], color="darkcyan", title="Avg Price by Flat Model (Top 10)")
axes[2,1].tick_params(axis="x", rotation=45)

lease_bins = pd.cut(
    df["remaining_lease_years"],
    bins=[0,20,40,60,80,100],
    labels=["0â20","21â40","41â60","61â80","81â100"]
)

df.assign(lease_group=lease_bins) \
  .groupby("lease_group")["resale_price"].mean() \
  .plot(kind="bar", ax=axes[2,2], color="goldenrod",
        title="Avg Price by Remaining Lease Group")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Scatter plots were used to examine the relationship between numerical variables and resale price. A positive correlation was observed between floor area and resale price, while remaining lease showed a clear price decay effect. These relationships justify the selection of these features for regression modelling

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 9))
fig.suptitle("Scatter Plot Analysis of Key Numerical Features", fontsize=14)

# Floor Area vs Resale Price
axes[0].scatter(
    df["floor_area_sqm"],
    df["resale_price"],
    color="steelblue"
)
axes[0].set_title("Floor Area vs Resale Price")
axes[0].set_xlabel("Floor Area (sqm)")
axes[0].set_ylabel("Resale Price ($)")

# Remaining Lease vs Resale Price
axes[1].scatter(
    df["remaining_lease_years"],
    df["resale_price"],
    color="darkorange"
)
axes[1].set_title("Remaining Lease vs Resale Price")
axes[1].set_xlabel("Remaining Lease (Years)")
axes[1].set_ylabel("Resale Price ($)")

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

## 5. Feature Selection

Based on EDA, the following features were selected for modelling due to their strong relationship with resale price:
- Floor area
- Remaining lease
- Flat age
- Flat type


The box plots illustrate the distribution and variability of resale prices. Significant differences in median prices and interquartile ranges are observed across flat types, with larger flats exhibiting higher medians and wider spreads. The presence of outliers highlights price variability in the resale market.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from tensorflow.keras.utils import to_categorical
```

```python
df["month"] = pd.to_datetime(df["month"])
df["year"] = df["month"].dt.year
df["flat_age"] = df["year"] - df["lease_commence_date"]

features = [
    "floor_area_sqm",
    "remaining_lease_years",
    "flat_age",
    "flat_type"  
]
X = df[features]
y = df['resale_price']
```

One-hot encoding was applied to convert categorical variables into numerical format suitable for regression modelling.

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

## 6. Multi-Linear Regression Model

Linear regression was implemented as a baseline model to establish initial predictive performance.


```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print('RÂ²: %.2f' %r2)
print("RMSE: %.2f" %rmse)
```

## Model performance 
RMSE was calculated by taking the square root of the mean squared error to ensure consistency with the unit of resale prices. On average, the modelâs predictions deviate from the actual resale price by approximately $113,212. Which indicates relatively high prediction error.

## To improve my RÂ² value
I included town as an additional features by using the average resale price by town. This is because transaction prices by town show strong relationship with resale prices, which significantly increased the model's RÂ² value.

```python
features = [
    "floor_area_sqm",
    "remaining_lease_years",
    "flat_age",
    "flat_type",
    "town",
]

X = df[features]
y = df["resale_price"]
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2_value = r2_score(y_test, y_pred)

print("Coefficients: %.2f" %model.coef_[0])
print("Intercept: %.2f" %model.intercept_)
print('RÂ²: %.2f' %r2_value)
print("RMSE: %.2f" %np.sqrt(mean_squared_error(y_test, y_pred)))

sample_flat = pd.DataFrame({
    "floor_area_sqm": [44],
    "remaining_lease_years": [61],
    "flat_age": [38],
    "flat_type": ["2 ROOM"],
    "town": ["ANG MO KIO"]
})

sample_encoded = pd.get_dummies(sample_flat)

# align columns with training data
sample_encoded = sample_encoded.reindex(
    columns=X_encoded.columns,
    fill_value=0
)

predicted_price = model.predict(sample_encoded)

print(f"Predicted resale price: ${predicted_price[0]:.0f}")

actual = 232000
predicted = predicted_price[0]

percent_error = abs(actual - predicted) / actual * 100
accuracy = 100 - percent_error

print(f"Percentage Error: {percent_error:.2f}%")
print(f"Approximate Accuracy: {accuracy:.2f}%")
```

## The predicted resale price is 192,626, while the actual resale price is $232,000. This results in a percentage error of approximately 17% for this individual prediction.
## On average, the predicted resale price deviates from the actual price by approximately 66,150. The error for this particular flat is within a reasonable range given the modelâs overall performance.

## 7. KNN Regression Model

I use KNN to compare whether multi-regression model is better or the KNN is better.
KNN regression was applied to capture non-linear relationships between housing features and resale price. Unlike linear regression, KNN predicts prices based on similarity between flats. Feature scaling was applied to ensure fair distance calculations. KNN achieved higher RÂ², indicating improved predictive performance.


```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

features = [
    "floor_area_sqm",
    "remaining_lease_years",
    "flat_age",
    "flat_type",
    "town",
]

X = df[features]
y = df["resale_price"]
```

```python
X_encoded = pd.get_dummies(X)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
for k in [3, 5, 7, 9, 11]:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f"k={k} â RÂ²: {r2:.2f}")
```

Choose K with the highest R2 Value 

```python
knn = KNeighborsRegressor(n_neighbors=7)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

r2_knn = r2_score(y_test, y_pred_knn)
rmse_knn = np.sqrt(mean_squared_error(y_test, y_pred_knn))

print(f"KNN RÂ²: {r2_knn:.2f}")
print(f"KNN RMSE: {rmse_knn:.2f}")
```

## 8. Model Comparison

## KNN regression significantly outperformed linear regression in predicting resale prices.
Why did KNN perform better then multi-regression? 
It is because housing prices are influenced by non-linear interactions, and KNN can model local similarity patterns that linear regression cannot.

Multiple values of k (5, 6, 7, and 11) achieved the same highest RÂ² value of 0.92. In such cases, a moderate k value was selected to balance predictive accuracy and model stability. A smaller k may be more sensitive to noise, while a larger k may overly smooth local patterns. Therefore, I choose k = 7 as it provides a good trade-off between robustness and the ability to capture local similarities among resale flats.

| Model               | RÂ²       | RMSE        | Remarks                                  |
| ------------------- | -------- | ----------- | ---------------------------------------- |
| Linear Regression   | 0.82     |   66154     | Baseline, limited by linear assumptions  |
| KNN Regressor (k=7) | **0.92** | **~43,802** | Best performance, captures non-linearity |


Linear regression achieved a moderate RÂ² value, indicating that a purely linear relationship was insufficient to model resale price variations. In contrast, KNN regression achieved a significantly higher RÂ², suggesting that resale prices are influenced by local similarity patterns and non-linear relationships. Therefore, KNN regression was selected as the final model.

## 9. Sample Prediction

A sample resale flat was used to demonstrate the prediction capability of the final model.

```python
sample_flat = pd.DataFrame({
    "floor_area_sqm": [44],
    "remaining_lease_years": [61],
    "flat_age": [38],
    "flat_type": ["2 ROOM"],
    "town": ["ANG MO KIO"]
})

sample_encoded = pd.get_dummies(sample_flat)
sample_encoded = sample_encoded.reindex(
    columns=X_encoded.columns, fill_value=0
)

sample_scaled = scaler.transform(sample_encoded)
pred_price = knn.predict(sample_scaled)

print(f"Predicted resale price: ${pred_price[0]:.2f}")

actual = 232000
predicted = pred_price[0]

percent_error = abs(actual - predicted) / actual * 100
accuracy = 100 - percent_error

print(f"Percentage Error: {percent_error:.2f}%")
print(f"Approximate Accuracy: {accuracy:.2f}%")
```

## The predicted resale price is 218,857, while the actual resale price is $232,000. This results in a percentage error of approximately 5%, which shows the model can produce reasonably accurate estimates for individual flats compare to multi-linear regression.
